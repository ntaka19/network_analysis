# Networkx trial 

```python
import networkx as nx
from matplotlib import pyplot as plt

Graph = nx.DiGraph()
Graph.add_node('hoge')
Graph.add_node('fuga')

Graph.nodes()
```

- まとめてノード追加

```python
Graph.add_nodes_from(['foo','bar'])
Graph.nodes(data=False)
```

- 属性付きのノードを作成する. 属性は好きなだけ追加することが可能。True/falseで属性の表示、非表示

```python
Graph.add_node('hoge', color = 'green')
Graph.add_node('fuga', color = 'red') 
Graph.nodes(data=True)
```

- ノードの削除

```python
Graph.nodes()
```

```python
Graph.remove_node('bar')
Graph.nodes()
```

- エッジの追加

```python
Graph.edges()
Graph.edges(data=False)
Graph.edges(data=True)
```

```python
Graph.nodes()
Graph.add_edges_from([('hoge','fuga'),('foo','bar')])
Graph.edges()

nx.draw_networkx(Graph)
plt.show()
```

## 2. Watts Strogatz Small World

```python
%matplotlib inline

import networkx as nx
import numpy as np
import matplotlib.pyplot as plt

#G= nx.DiGraph()

G = nx.watts_strogatz_graph(n=100,k=5,p=0.75)


nx.draw_networkx(G)
plt.show()


#print(G.edges())
print(nx.shortest_path(G)[0][1])

print(nx.average_shortest_path_length(G))

print(np.average(nx.degree(G).values()))
```
